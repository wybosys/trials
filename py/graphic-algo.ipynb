```python
# 图论包测试

import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.DiGraph()
G.add_node('a')
G.add_node('a')
G.add_node('b')
G.add_node('c')
G.add_node('d')
G.add_edge('a', 'c')
G.add_edge('b', 'c')
G.add_edge('b', 'd')
```

```python
nx.draw_networkx(G, nx.spring_layout(G))
```

```python
dag = list(nx.dag.topological_sort(G))
print(dag)
```
